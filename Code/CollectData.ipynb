### Lets pull all the Fracture and Edema Images to its repective folder

```python
# we are dealing with SAGIR images
img = 'SAGIR'

def prepareData(imgName, img):
    """
    Function to copy the images from main data folder to
    a different folder for all img images
    """
    
    # initialize the counters to track the number of copies
    unhealthy_counter = 0
    unhealthy_frac_count = 0
    unhealthy_edema_count = 0
    healthy_counter = 0

    print(f"Total unique unhealty images are :::: {len(imgName)}")
    # images having Edema
    edema_img_names = df_bounding[df_bounding['value_rectanglelabels'] == '[\'Edema\']']['image'].unique()
    print(f"Total  unique Images that has edema are ::: {len(edema_img_names)}")
    # images having Fracture
    frac_img_names = df_bounding[df_bounding['value_rectanglelabels'] == '[\'Fracture\']']['image'].unique()
    print(f"Total  unique Images that has fracture are ::: {len(frac_img_names)}")

    print("Start of image seggregation to respective folders......")

    # go through the images and segregate
    for root, dirs, files in os.walk(image_dir):
        for file_name in files:
            # ignore the store file
            if file_name != '.DS_Store':
                file_path = os.path.join(root, file_name)
                # if it is a SAGIR image 
                if img in file_path:
                    # take the unhealty images to differnt folder for training and testing
                    if file_name in imgName:
                        # if the file has edema copy it to the edema folder 
                        # and if it is also fracture copyto the fracture folder
                        if file_name in edema_img_names:
                            shutil.copy(file_path, dest_edem_SAGIR)
                            unhealthy_edema_count += 1
                        if file_name in frac_img_names:
                            shutil.copy(file_path, dest_frac_SAGIR)
                            unhealthy_frac_count += 1
                        unhealthy_counter += 1

                    if file_name not in imgName:
                        # copy only those healthy image slice which has more than 65% of the image as foot
                        h_img = Image.open(file_path)
                        h_img_gray = h_img.convert("L")
                        h_img_array = np.array(h_img_gray)
                        h_size = h_img.size
                        black_pix = np.sum(h_img_array <= 10) # pixel value <=10 considered as black

                        if( (black_pix / (h_size[0] * h_size[1])) < 0.65 ):
                            shutil.copy(file_path, healthy_SAGIR)
                            healthy_counter +=1
    
    print("Total Unhealthy Images copied edema/fracture::", unhealthy_counter)
    print("Total Unhealthy Images having fracture::", unhealthy_frac_count)
    print("Total Unhealthy Images having edema::", unhealthy_edema_count)
    print("Total healthy Images copied::", healthy_counter)
```

```python
# prepare the data
prepareData(imgName, img)
```
